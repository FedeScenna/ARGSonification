### Argentina Sonification

This project is about attempting to *hear* Argentine economic data. There is a wide variety of cool visualizations out there. However, other forms of representing data other than visually have emerged.

In this notebook, I will download argentine economic activity time series data and attempt to create a sonification of it. However, I will previously visualize it so compare how sound gets mapped.

#### 1 - Data download

Argentine government has created an API so the public can access government data more easily. Economic activity data is published by the National Institute of Statistics and Census (INDEC)[https://www.indec.gob.ar/]. However, the format is not quite tidy and requires loads of manual transformations. Getting the data from the API saves a considerable amount of work. 

```python
import pandas as pd
import requests
import urllib.parse

def get_api_call(ids, **kwargs):
    API_BASE_URL = "https://apis.datos.gob.ar/series/api/"
    kwargs["ids"] = ",".join(ids)
    return "{}{}?{}".format(API_BASE_URL, "series", urllib.parse.urlencode(kwargs))
```

```python
df = pd.read_csv(get_api_call(
    ["143.3_NO_PR_2004_A_21:percent_change"],
    format="csv", start_date=2004
))
```
